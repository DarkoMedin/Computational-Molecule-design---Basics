For this exemplar tutorial, pandas and numpy will be required for basic Python operations and rdkit will be used to deploy data and chemoinformatics procedures to made molecues machine readable and plot them.

```python
#Import the required libraries
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem
```

For start, i will show to use a simple notation called SMILES or Simplified molecular-input line-entry system which will not only simplifify how we can present molecules but also make this notation readable for Machines or Machine Learning algorhitms.

CH3CH2CH2CH2CH3 or pentane would be CCCCCC as simple as that just 5 C atoms and the hydrogen attomes are assumed

```python
pentane='CCCCC'
```

```python
#Add the basic pentane molecular structure using the  SMILES notation directly
molstr = Chem.MolFromSmiles(pentane)
molstrH=Chem.AddHs(molstr)
molstrH
```

Same molecular structure can be created using the SMILE notation directly between the brackets of the MolFromSmiles()

```python
#Add the basic pentane molecular structure using the object containing SMILES notation as a string
molstr = Chem.MolFromSmiles("CCCCC")
molstrH=Chem.AddHs(molstr)
molstrH
```

A very simple way to add the double bond is using "=" sign and i will position it between second and third carbon atoms

```python
#Add the double bond at the position 2
molstr2 = Chem.MolFromSmiles("CC=CCC")
molstrH2=Chem.AddHs(molstr2)
molstrH2
```

Adding the branch - Along with the double bond on position 1 i will now specify branching  carbon atom on position 5. Remeber the
naming principles of organic molecules are based on double bonds if present

```python
molstr3 = Chem.MolFromSmiles("C=CCC(C)CCC")
molstrH3=Chem.AddHs(molstr3)
molstrH3
```

The next task i would like to add here is creating the circular - aromatic molecules, essential for organic molecule design, present in most essential biological molecules such as DNA, RNA, amino acids, hormones and many others. The SMILES for simple hexagonal aromatic molecule hexane would be achieved by specifying the start and end molecules where closing the aromatic structure is to be performed. This is achieved simply by adding 1 on the first C and another 1 on the last C.

```python
molstr5 = Chem.MolFromSmiles("C1CCCCC1")
molstrH5=Chem.AddHs(molstr5)
molstrH5
```

Now if i wanted to add chlorine for example to the last carbon atom i would just add "Cl" next to the last C. For simplicity 
and better visualization i will not add H atoms into visualizations, but keep in mind that they are assumed.

```python
molstr6 = Chem.MolFromSmiles("C1CCCCC1Cl")
molstr6
```

Similarly another metal atom can be added for example on the first C1 - "Br"

```python
molstr7 = Chem.MolFromSmiles("BrC1CCCCC1Cl")
molstr7
```

The next task - Add the rotating double bonds inside of the aromatic ring and create a benzene or cyclohexene.
This is done very simply by lowering the letters of all Cs into cs and the Chem.MolFromSmiles will read them as cyclohexene with 
Cl and Br being added to the aromatic ring.

```python
molstr8 = Chem.MolFromSmiles("Brc1ccccc1Cl")
molstr8
```

The next task - Combine the aromatic ring with the linear 1-pentene. This can easily be achieved using SMILES by addidng C=CCCC to c1, but keep in ming that that C needs to have free hydrogens, the c1 attom bound with Cl 

```python
molstr9 = Chem.MolFromSmiles("C=CCCCc1ccccc1Cl")
molstr9
```

Adding OH group vsn be achieved by adding simple "O" at the the desired position, in this case linear end of the molecule.

```python
molstr10 = Chem.MolFromSmiles("OC=CCCCc1ccccc1Cl")
molstr10
```

The task - How to add whole benzene ring using the "-OH" group. Adding the "c1ccccc1" next to the "O" atom will automatcally mean 
that O molecule now is the link to a new benzed ring added.

```python
molstr11 = Chem.MolFromSmiles("c1ccccc1OC=CCCCc1ccccc1Cl")
molstr11
```

New task - Now i will create the Lysine molecule using SMILES  and plot it.
Lysine a NH2 group one end and COOH group on the other end. Additionaly, another NH2 group is located next to the COOH containing 
carbon atom. The NH2 groups are added very simply, using the "N" notation while -COOH group is noted as "C(=O)O"

```python

molstr12 = Chem.MolFromSmiles("NCCCCC(C(=O)O)N")
molstr12
```

Final thoughts - What was covered in this tutorial are the SMILES notations which can be used to make the
